```python
data = [[11, 12, 13, 14], [21, 22, 23, 24], [31,32,33,34]]
data
```

```python
from xlwt import Workbook
book = Workbook()
sheet1 = book.add_sheet("Data Book1")
row = 0
for line in data:
    col = 0
    for datum in line:
        sheet1.write(row, col, datum)
        col += 1
    row += 1
```
