```python
# -*- coding: utf-8 -*-

import sqlite3
from datetime import datetime
import time
```

```python
connect = sqlite3.connect('InfoStock.db')
cursor = connect.cursor()
```

Query all items in a table INFORSTOCK

```python
cursor.execute('SELECT * FROM INFOSTOCK ')
for row in cursor:
    print ('%s %s %s' % ( row[0], row[1], row[2]) )
```

Query with Date (2015-08-27일만 찾기)

```python
cursor.execute("SELECT * FROM INFOSTOCK WHERE date BETWEEN date('2015-08-27') AND date('2015-08-28')")
for row in cursor:
    print ('%s %s' % ( row[0], row[2]) )
```

2015-08-27 9시-13시

```python
cursor.execute("SELECT * FROM INFOSTOCK WHERE date >= '2015-08-28 09:00' AND date < '2015-08-28 13:00'")
for row in cursor:
    print ('%s %s' % ( row[0], row[2]) )
```

title에서 '증시'만 찾기

```python
cursor.execute("SELECT * FROM INFOSTOCK where title like '%기술적 분석 특징주 A%' AND title like '%증시요약%' ")
for row in cursor:
    print ('%s %s %s' % ( row[0], row[1],row[2]) )
```
